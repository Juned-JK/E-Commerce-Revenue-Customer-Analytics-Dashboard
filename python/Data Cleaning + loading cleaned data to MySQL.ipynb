```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\juned\Videos\data analytics\end to end projects\Ecommerce\ecommerce.csv",encoding='latin1')
```

# Data cleaning 

```python
df
```

```python
df['Revenue'] = df['Quantity'] * df['UnitPrice'] #created Revenue column
```

```python
df
```

```python
df[df['InvoiceNo'].str.contains('C')] # finding cancelled orders
```

```python
df = df[~df['InvoiceNo'].astype(str).str.startswith('C', na=False)].copy() # Removing cancelled orders
```

```python
df[df['InvoiceNo'].str.contains('C')] # cancelled orders are removed 
```

```python
df.drop_duplicates()
```

```python
df.dropna(subset=['CustomerID'], inplace = True) # removing blank CustomerID
```

```python
df
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate']) # converting str column to date time 

df['Day'] = df['InvoiceDate'].dt.day #creating Day column
df['Month'] = df['InvoiceDate'].dt.month #creating month column
df['Year'] = df['InvoiceDate'].dt.year #creating year column
```

```python
df
```

```python
df.drop(columns=['InvoiceDate'], inplace = True) # removing InvoiceDate column
```

```python
df
```

```python
# Creating a new Colume
df.loc[df['Month'].isin([1,2,3]), 'Quarter'] = 'Q1'
df.loc[df['Month'].isin([4,5,6]), 'Quarter'] = 'Q2'
df.loc[df['Month'].isin([7,8,9]), 'Quarter'] = 'Q3'
df.loc[df['Month'].isin([10,11,12]), 'Quarter'] = 'Q4'
```

```python
df
```

```python
df.fillna(' ') # it will remove NaN values with a blank space
```

```python
df.dropna() # it will drop all the column that contains blank values
```

```python
#storing cleaned data in Excel format
df.to_excel(r"C:\Users\juned\Videos\data analytics\end to end projects\Ecommerce\Cleaned_Ecommerce_data.xlsx",index=False) 
```

```python
#storing cleaned data in CSV
df.to_csv(r"C:\Users\juned\Videos\data analytics\end to end projects\Ecommerce\Cleaned_Ecommerce_data.csv",index=False) 
```
